```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Load the data
data = pd.read_csv('VennRMF_ALL.csv', delimiter=';')

# Create a boxplot
sns.boxplot(data=data)
plt.xlabel('Sourdough samples')
plt.ylabel('Prvalence of microbes')

# Calculate the p-value
p_value = f_oneway(data['RS'], data['MS'], data['FS']).pvalue

# Add the p-value to the plot
plt.annotate(f'p-value = {p_value:.3f}', xy=(0.05, 0.96), xycoords='axes fraction')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Load the data
data = pd.read_csv('VennRMF_ALL.csv', delimiter=';')

# Set the color palette
colors = ['#ffa600', '#ff6361', '#58508d']

# Create a boxplot
sns.boxplot(data=data, palette=colors)
plt.xlabel('Sourdough samples', fontsize=10)
plt.ylabel('Prevalence of microbes', fontsize=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Microbial prevalence in different sourdough samples', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Calculate the p-value
p_value = f_oneway(data['RS'], data['MS'], data['FS']).pvalue

# Add the p-value to the plot
plt.annotate(f'p-value = {p_value:.3f}', xy=(0.75, 0.96), xycoords='axes fraction', fontsize=10, color='red')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('VennRMF_ALL.csv', delimiter=';')

# Calculate the number of species found in each type of sourdough
rs_count = data['RS'].sum()
ms_count = data['MS'].sum()
fs_count = data['FS'].sum()

# Create a bar chart to visualize the data
plt.bar(['Rye sourdough', 'Mesophilic sourdough', 'Fermented sourdough'], [rs_count, ms_count, fs_count])
plt.title('Number of species found in each type of sourdough')
plt.xlabel('Type of sourdough')
plt.ylabel('Number of species')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load data from CSV file
data = pd.read_csv('VennRMF_ALL.csv', delimiter=';')

# Create a contingency table of the data
contingency_table = [
    [data['RS'].sum(), data['MS'].sum(), data['FS'].sum()],
    [data[data['RS'] > 0]['RS'].count(), data[data['MS'] > 0]['MS'].count(), data[data['FS'] > 0]['FS'].count()]
]

# Perform chi-square test on the contingency table
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f'Chi-square value: {chi2}')
print(f'P-value: {p_value}')
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Load data from CSV
data = pd.read_csv("VennRMF_ALL.csv", sep=";")

# Convert count data to binary data
y = data[['RS', 'MS', 'FS']].applymap(lambda x: 1 if x > 0 else 0)

# Create logistic regression model
model = LogisticRegression(random_state=0)

# Fit the model
model.fit(y, data['NAME'])

# Print the coefficients
print(model.coef_)
```

```python
import pandas as pd
import seaborn as sns

# Load data from CSV
data = pd.read_csv("VennRMF_ALL.csv", sep=";")

# Convert count data to binary data
binary_data = data[['RS', 'MS', 'FS']].applymap(lambda x: 1 if x > 0 else 0)

# Combine data into one dataframe
combined_data = pd.concat([data['NAME'], binary_data], axis=1)

# Pivot data
pivoted_data = combined_data.melt(id_vars=['NAME'], value_vars=['RS', 'MS', 'FS'], var_name='Sourdough Type', value_name='Presence')

# Create heatmap
sns.heatmap(pivoted_data.pivot(index='NAME', columns='Sourdough Type', values='Presence'), cmap='YlGnBu')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file
data = pd.read_csv('/workspaces/datathinking.org-codespace/data/VennRMF_ALL.csv', delimiter=';')

# Create a horizontal bar plot
fig, ax = plt.subplots(figsize=(18, 16))

# Color code the bars based on sample type
colors = {'RS': 'r', 'MS': 'g', 'FS': 'b'}

for i, col in enumerate(['RS', 'MS', 'FS']):
    ax.barh(data['NAME'], data[col], color=colors[col], label=col)

# Add a legend
ax.legend(loc='center right', fontsize=7)

# Set the title and axis labels
ax.set_title('Microbial Species in Rye, Mesophilic, and Fermented Sourdough', fontsize=16)
ax.set_xlabel('Number of Samples', fontsize=10)
ax.set_ylabel('Microbial Species', fontsize=10)

# Set the tick label font size
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)

# Add gridlines
ax.grid(True)

# Adjust the figure margins
fig.subplots_adjust(left=0.25, bottom=0.1)

plt.show()
```
